```python
mydict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(mydict)
print(mydict["brand"])
print(mydict.get("model"))
print(mydict.keys())
print(mydict.values())
print(mydict.items())

mydict.pop("model")
print(mydict)

del(mydict["year"]) #func type
print(mydict)

mydict.popitem()  #delete the last item
print(mydict)

mydict.clear()
print(mydict)
```

```python
myfamily = {
    "Child1": {
        "name": "Morbius",
        "year": 2001
    },
    "Child2": {
        "name": "Lucifer",
        "year": 2007
    },
    "Child3": {
        "name": "Ezio",
        "year": 2011
    }
}
print(myfamily["Child2"]["name"])

myfamily["Child2"]["name"] = "Lucifer Morningstar"
print(myfamily["Child2"]["name"])

del(myfamily["Child3"]["name"])
print(myfamily)

myfamily["Child3"].pop("year")
print(myfamily)
```

```python
def num(start, end):
  even = []
  for n in range(start, end+1):
    if n % 2 == 0:
      even.append(n)
  return even

s = int(input("Enter start number: "))
e = int(input("Enter end number: "))
print(num(s, e))
```
