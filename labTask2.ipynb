```python
count = 10
num = 20

while count > 0:
    print(num)
    num -= 2
    count -= 1
```

```python
n = int(input("Enter a range: "))
fact = 1
for i in range(1, n):
    fact *= i
    print(fact)
print("Factorial of 5 is:", fact)
```
